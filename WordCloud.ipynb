```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
from string import punctuation
import matplotlib.pyplot as plt
```

```python
olx_electronics = pd.read_csv('olx_electronics.csv')
olx_fashion = pd.read_csv('olx_fashion.csv')
olx_realestate = pd.read_csv('olx_realestate.csv')
olx_vehicles = pd.read_csv('olx_vehicles.csv')
```

```python
e = ""
for i in range(0,len(olx_electronics)):
    e += str(olx_electronics.ix[i]['desc'])
    
f = ""
for i in range(0,len(olx_electronics)):
    f += str(olx_fashion.ix[i]['desc'])
    
r = ""
for i in range(0,len(olx_electronics)):
    r += str(olx_realestate.ix[i]['desc'])
    
v = ""
for i in range(0,len(olx_electronics)):
    v += str(olx_vehicles.ix[i]['desc'])
```

```python
e = e.lower()
f = f.lower()
r = r.lower()
v = v.lower()
```

```python
def strip_punctuation(s):
    return ''.join(c for c in s if c not in punctuation)
```

```python
strip_punctuation(e)
strip_punctuation(f)
strip_punctuation(r)
strip_punctuation(v)
```

```python
eCloud = WordCloud(stopwords=STOPWORDS, background_color='black', width=1800, height=1400).generate(e)
fCloud = WordCloud(stopwords=STOPWORDS, background_color='black', width=1800, height=1400).generate(f)
rCloud = WordCloud(stopwords=STOPWORDS, background_color='black', width=1800, height=1400).generate(r)
vCloud = WordCloud(stopwords=STOPWORDS, background_color='black', width=1800, height=1400).generate(v)
```
